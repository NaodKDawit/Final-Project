```r
#Remove unnecessary columns

df <- subset(df, select = -c(Timestamp, Permissions))

#Find and remove missing values
colSums(is.na(df))
df[df ==""] <-NA
df_1 <- na.omit(df)
print(df_1)
colSums(is.na(df_1))
summary(df_1)
```

```r
# Calculate the IQR for Hours per day
Q1 <- quantile(df_1$Hours.per.day, 0.25, na.rm = TRUE)
Q3 <- quantile(df_1$Hours.per.day, 0.75, na.rm = TRUE)
IQR_hours <- IQR(df_1$Hours.per.day, na.rm = TRUE)
low_hours <- Q1 - 1.5 * IQR_hours
up_hours <- Q3 + 1.5 * IQR_hours

# Calculate the IQR for BPM
Q1_1 <- quantile(df_1$BPM, 0.25, na.rm = TRUE)
Q3_1 <- quantile(df_1$BPM, 0.75, na.rm = TRUE)
IQR_bpm <- IQR(df_1$BPM, na.rm = TRUE)
low_bpm <- Q1_1 - 1.5 * IQR_bpm
up_bpm <- Q3_1 + 1.5 * IQR_bpm
```

```r
# Add filtered data to new data set
df_clean <- df_1 %>% filter(Hours.per.day >= low_hours & Hours.per.day <= up_hours) %>%
  filter(BPM >= low_bpm & BPM <= up_bpm)

sapply(df_clean, class)

# Mutating characters as factors (music effects)
df_clean <- df_clean %>%
  mutate(across(where(is.character), ~na_if(.x, ""))) %>%
  mutate(across(where(is.factor), ~na_if(as.character(.x), ""))) %>%
  mutate(Music.effects = dplyr::recode(Music.effects,`0` = "Worsen", `1` = "No Effect",`2` = "Improve")
  ) %>%
  mutate(across(c("Music.effects"), as.factor))
```

```r
# Summary Stats of Continuous Variables
summary_cont <- df_clean %>%
  select(Age, Hours.per.day, BPM) %>%
  summarise(across(everything(), list(
    Mean = ~mean(.x, na.rm = TRUE),
    SD = ~sd(.x, na.rm = TRUE),
    Min = ~min(.x, na.rm = TRUE),
    Max = ~max(.x, na.rm = TRUE),
    NonMissing = ~sum(!is.na(.x)),
    Missing = ~sum(is.na(.x))
  ), .names = "{.col}_{.fn}")) %>%
  pivot_longer(everything(), names_to = c("Variable", ".value"), names_sep = "_") %>%
  select(Variable, Mean, SD, Min, Max, NonMissing, Missing)

print(summary_cont)

# Summary Stats of Categorical Variables
cat_vars <- c("Depression_binary", "Music.effects", "Fav.genre")

cat_summary <- map_dfr(cat_vars, function(var) {
  x <- df_clean[[var]]
  freq <- table(x, useNA = "no")
  levels_cat <- paste(names(freq), "=", as.integer(freq), collapse = ", ")
  tibble(
    Variable = var,
    Levels = n_distinct(x, na.rm = TRUE),
    Counts = levels_cat,
    NonMissing = sum(!is.na(x)),
    Missing = sum(is.na(x))
  )
})

print(cat_summary)
```

```r
# Turning Music effects into binary for improve
df_clean <- df_clean %>%
  mutate(MusicImprove = ifelse(Music.effects == "Improve", 1, 0))

# Logistic model for Music Improve
logit_model <- glm(MusicImprove ~ Depression + BPM + Hours.per.day + Age,
                   data = df_clean, family = binomial)
summary(logit_model)
exp(coef(logit_model))  # Odds ratios
exp(confint(logit_model))    # Confidence intervals for Odds ratios

# Remove intercept for analysis
tidy_m <- tidy(logit_model, conf.int = TRUE, exponentiate = TRUE) %>%
  filter(term != "(Intercept)")

# Plot with error bars
ggplot(tidy_m, aes(x = term, y = estimate)) +
  geom_point() +
  geom_errorbar(aes(ymin = conf.low, ymax = conf.high), width = 0.2, col = "red") +
  geom_text(aes(label = paste0("(", round(conf.low, 4), ", ", round(conf.high, 4), ")")), vjust = -1.9) +
  coord_flip() + theme_minimal() + labs(title = " Odds Ratio Estimates with 95% CIs for Logistic Regression", y = "Estimate", x = "Predictor")


# Logistic model for Depression Binary
logit_depression <- glm(Depression_binary ~ BPM + Hours.per.day + Age + Music.effects,
                        data = df_clean, family = binomial)
summary(logit_depression)
exp(coef(logit_depression))  # Odds ratios
exp(confint(logit_depression))    # Confidence intervals for Odds ratios

# Remove intercept for analysis
tidy_d <- tidy(logit_depression, conf.int = TRUE, exponentiate = TRUE) %>%
  filter(term != "(Intercept)")

# Plot with error bars
ggplot(tidy_d, aes(x = term, y = estimate)) +
  geom_point() +
  geom_errorbar(aes(ymin = conf.low, ymax = conf.high), width = 0.3, col = "red") +
  geom_text(aes(label = paste0("(", round(conf.low, 4), ", ", round(conf.high, 4), ")")), vjust = -1.9, hjust = -0.1) +
  coord_flip() + theme_minimal() + labs(title = " Odds Ratio Estimates with 95% CIs for Logistic Regression", y = "Estimate", x = "Predictor")
```
